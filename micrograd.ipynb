```python
import math 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
def f(x): 
    return 3*x**2 - 4*x + 5; 
```

```python
f(3.0)
```

```python
xs = np.arange(-5, 5, 0.25) # generate a range of values with a step of 0.25
xs
```

```python
ys = f(xs)
ys
plt.plot(xs, ys)
plt.show()
```

```python
# finding the derivative of the function: 
# definition of the derivative: 
# f'(x) = lim h->0 (f(x+h) - f(x)) / h
epsilon = 1e-10
def df(x):
    # using newton methods: 
    print("Epsilon: ", epsilon)
    print(f(x+epsilon) - f(x))
    print ((f(x+epsilon) - f(x)) / epsilon)
    
ys_differentiable = df(3.0)


xs = np.arange(-5, 5, 0.25)
ys_differentiable = df(xs)
plt.plot(xs, ys_differentiable)
plt.show()
```
